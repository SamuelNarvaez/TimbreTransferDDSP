```python
import os

from pydub import AudioSegment
files = os.listdir('/home/ubuntu/DagstuhlChoirSet/audio_wav_22050_mono')
j=0
base = '/home/ubuntu/DagstuhlChoirSet/audio_wav_22050_mono/'
dropped = []
for i in files:
    i = base + i
    try:
        j+=1
        print(i, (100*j)/1658,'% done')
        sound = AudioSegment.from_wav(i)
        sound = sound.set_channels(1)
        sound = sound.set_frame_rate(16000)
        sound.export(i[:-3]+'wav', format="wav")
    except:
        dropped.append(i)
        os.remove(i)
```
